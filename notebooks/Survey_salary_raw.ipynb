```python
import pandas as pd

# Missing report
def missing_report(df):
    return pd.DataFrame({
        'Missing': df.isnull().sum(),
        'Percent': (df.isnull().mean() * 100).round(2)
    })

print(missing_report(df))
```

1.2Handle missing

```python
# salary flag
df['salary_imputed'] = df['annual_salary'].isnull().astype(int)

# convert salary to numeric
df['annual_salary'] = (
    df['annual_salary']
    .astype(str)
    .str.replace(',', '')
)

df['annual_salary'] = pd.to_numeric(df['annual_salary'], errors='coerce')

# fill median theo industry
industry_median = df.groupby('industry')['annual_salary'].transform('median')

df['annual_salary'] = df['annual_salary'].fillna(industry_median)
print(df['annual_salary'].isnull().sum())
```

1.3 Remove Duplicated


```python
print(df.duplicated().sum())

df = df.drop_duplicates()
```

1.4 Fix datatype

```python
# salary về numeric
df = df.copy()
df['annual_salary'] = (
    df['annual_salary']
    .astype(str)
    .str.replace(',', '')
)

df['annual_salary'] = pd.to_numeric(df['annual_salary'], errors='coerce')
print(df['annual_salary'].head())
```
