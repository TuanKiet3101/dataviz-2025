```python
groupby()
Dùng khi nào

Dùng để nhóm dữ liệu theo cột và tính toán thống kê.
Syntax
df.groupby("column")
```

```python
import pandas as pd

data = {
    "Department": ["IT", "IT", "HR", "Finance"],
    "Salary": [1000, 1200, 900, 1500]
}

df = pd.DataFrame(data)

result = df.groupby("Department")["Salary"].mean()

print(result)
```

```python
merge()
Dùng khi nào

Dùng để nối 2 DataFrame giống SQL JOIN.

Syntax
pd.merge(df1, df2, on="column")
```

```python
students = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["An", "Binh", "Cuong"]
})

scores = pd.DataFrame({
    "ID": [1, 2, 3],
    "Score": [8, 9, 7]
})

result = pd.merge(students, scores, on="ID")

print(result)
```

```python
pivot_table()
Dùng khi nào

Tạo bảng tổng hợp dữ liệu giống Pivot Table trong Excel.
Syntax 
df.pivot_table(values, index, columns)
```

```python
sales = pd.DataFrame({
    "Month": ["Jan", "Jan", "Feb", "Feb"],
    "Product": ["A", "B", "A", "B"],
    "Revenue": [100, 150, 200, 250]
})

result = sales.pivot_table(
    values="Revenue",
    index="Month",
    columns="Product",
    aggfunc="sum"
)

print(result)
```

```python
apply()
Dùng khi nào

Áp dụng function lên cột hoặc hàng.

Syntax
df["column"].apply(function)
```

```python
df = pd.DataFrame({
    "Name": ["Lan", "Minh", "Hoa"],
    "Score": [7, 8, 9]
})

df["Double Score"] = df["Score"].apply(lambda x: x * 2)

print(df)
```

```python
melt()
Dùng khi nào

Chuyển dữ liệu từ wide format sang long format.

Syntax
pd.melt(df)
```

```python
df = pd.DataFrame({
    "Student": ["An", "Binh"],
    "Math": [8, 9],
    "English": [7, 8]
})

result = pd.melt(
    df,
    id_vars=["Student"],
    value_vars=["Math", "English"],
    var_name="Subject",
    value_name="Score"
)

print(result)
```

```python
dropna()
Dùng khi nào

Xóa dữ liệu bị thiếu (NaN).

Syntax
df.dropna()
```

```python
df = pd.DataFrame({
    "Name": ["An", None, "Cuong"],
    "Score": [8, 9, None]
})

print(df.dropna())
```

```python
sort_values()
Dùng khi nào

Sắp xếp dữ liệu theo cột.

Syntax
df.sort_values(by="column")
```

```python
df = pd.DataFrame({
    "Name": ["Lan", "Minh", "Hoa"],
    "Score": [7, 9, 8]
})

result = df.sort_values(by="Score", ascending=False)

print(result)
```

```python
fillna()
Dùng khi nào

Điền dữ liệu bị thiếu.

Syntax
df.fillna(value)
```

```python
df = pd.DataFrame({
    "Name": ["An", "Binh", None],
    "Score": [8, None, 7]
})

result = df.fillna(0)

print(result)
```

```python
np.mean()
Dùng khi nào

Tính giá trị trung bình.

Syntax
np.mean(array)
```

```python
import numpy as np

arr = np.array([2, 4, 6, 8])

result = np.mean(arr)

print(result)
```

```python
arr = np.array([10, 25, 5, 40])

result = np.max(arr)

print(result)
```
