```python
#function Groupby 
#groupby() dùng để:nhóm dữ liệu theo cột,tính toán như sum, mean, count
#các Parameters thường dùng bao gồm 
#| Parameter  | Ý nghĩa                         
 #`by`       | cột dùng để nhóm                 
# `axis`     | nhóm theo hàng/cột               
# `as_index` | có dùng cột nhóm làm index không 
# `sort`     | có sắp xếp nhóm không            

```

```python
import pandas as pd

data = {
    "Department": ["IT", "IT", "HR", "HR", "Finance"],
    "Salary": [1000, 1200, 900, 950, 1500]
}

df = pd.DataFrame(data)

result = df.groupby("Department")["Salary"].mean()

print(result)
```

```python
#Merge
#Dùng để nối 2 DataFrame giống JOIN trong SQL.
# Parameter | Ý nghĩa                                      
# `left`    | bảng trái                                    
# `right`   | bảng phải                                    
# `on`      | cột chung                                    
# `how`     | kiểu nối (`inner`, `left`, `right`, `outer`) 
```

```python
students = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["An", "Binh", "Cuong"]
})

scores = pd.DataFrame({
    "ID": [1, 2, 3],
    "Score": [8, 9, 7]
})

result = pd.merge(students, scores, on="ID")

print(result)
```

```python
#Pivot_table
#Dùng để tạo bảng tổng hợp giống Excel Pivot Table 
#| Parameter | Ý nghĩa       
#| `values`  | cột tính toán 
#| `index`   | hàng          
#| `aggfunc` | hàm tổng hợp  
```

```python
sales = pd.DataFrame({
    "Month": ["Jan", "Jan", "Feb", "Feb"],
    "Product": ["A", "B", "A", "B"],
    "Revenue": [100, 150, 200, 250]
})

result = sales.pivot_table(
    values="Revenue",
    index="Month",
    columns="Product",
    aggfunc="sum"
)

print(result)
```

```python
#Apply 
#Thường dùng để áp dụng function lên cột hoặc hàng
#| Parameter | Ý nghĩa           |
#| `func`    | hàm áp dụng       |
#| `axis`    | 0 = cột, 1 = hàng |
```

```python
df = pd.DataFrame({
    "Name": ["Lan", "Minh", "Hoa"],
    "Score": [7, 8, 9]
})

df["Score x2"] = df["Score"].apply(lambda x: x * 2)

print(df)
```

```python
#melt()
#Dùng để chuyển dữ liệu từ dạng rộng ra dàng dài
#| Parameter    | Ý nghĩa                   |
#| `id_vars`    | cột giữ nguyên            |
#| `value_vars` | cột chuyển đổi            |
#| `var_name`   | tên cột mới chứa tên biến |
#| `value_name` | tên cột giá trị           |
```

```python
df = pd.DataFrame({
    "Student": ["An", "Binh"],
    "Math": [8, 9],
    "English": [7, 8]
})

result = pd.melt(
    df,
    id_vars=["Student"],
    value_vars=["Math", "English"],
    var_name="Subject",
    value_name="Score"
)

print(result)
```
